```python
##Q7
df['emp_len_numbers'] = df['emp_length'].astype('str').str.extractall('(\d+)').unstack().fillna('').sum(axis=1).astype(int)
df['emp_len_numbers']

```

```python
##q8
import re as re
def findnum(text):
    num = re.findall(r'[0-9]+',text)
    #print(num)
    return " ".join(num)
df['term_num']=df['term'].apply(lambda x: findnum(x))
df['term_num']
```

```python
##q10

import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize = (20,20))
sns.countplot(data = df,x = df["loan_status"],hue = df["grade"])
plt.show()
```

```python
##q11
''' 11.Using a user defined function convert the ‘emp_len’ column into
categorical column as follows -
If emp_len is less than equals to 1 then recode as ‘fresher’.
If emp_len is greater than 1 and less than 3 then recode as ‘junior’.
If emp_len is greater than 3 and less than 7 then recode as ‘senior’
If emp_len is greater than 7 then recode as ‘expert’.'''

df['emp_len_numbers']
def catog(x):
    if x<=1:
        return 'fresher'
    elif x>1 and x<3:
        return 'junior'
    elif x>3 and x<7:
        return 'senior'
    elif x>7:
        return 'expert'

df['emp_len_category']=df['emp_len_numbers'].map(catog)    

df['emp_len_category']
```

```python
##12

loan=pd.pivot_table(df,index='grade',values='loan_amnt',aggfunc='sum')
print(loan)
plt.pie(x=loan.loan_amnt,labels=loan.index,autopct='%2f',radius=2.5)
plt.show()
```
